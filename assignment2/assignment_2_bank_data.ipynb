```python
encoder = ce.OneHotEncoder(cols=np.array(X.columns))
X_o = encoder.fit_transform(X)

X_o
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_o, y, test_size=0.25, random_state=0)
```

# Naive Bayes Algorithm

```python
## oversample the 'yes' class so that both classes have the same number of data points
oversample = RandomOverSampler(sampling_strategy=0.5)
X_overtrain, y_overtrain = oversample.fit_resample(X_train, y_train)

cnb = CategoricalNB(alpha =1, min_categories = 2)

y_pred = pd.Series( cnb.fit(X_overtrain, y_overtrain).predict_proba(X_test)[:,1]>0.0001
                  ).replace({True: 'yes', False: 'no'})
print('Test-set accuracy score: {0:0.5f}'. format(accuracy_score(y_test, y_pred)))
print('Test-set recall score: {0:0.5f}'. format(recall_score(y_test, y_pred, pos_label='yes')))
```

```python
y_pred_train = pd.Series(cnb.predict_proba(X_train)[:,1]>0.00005
                  ).replace({True: 'yes', False: 'no'})
print('Training-set accuracy score: {0:0.5f}'. format(accuracy_score(y_train, y_pred_train)))
print('Training-set recall score: {0:0.5f}'. format(recall_score(y_train, y_pred_train, pos_label='yes') ))
```

```python
cm  = confusion_matrix(y_test, y_pred)
TP = cm[1,1]
TN = cm[0,0]
FP = cm[0,1]
FN = cm[1,0]



print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[1,1])

print('\nTrue Negatives(TN) = ', cm[0,0])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])

print(classification_report(y_test, y_pred))


false_positive_rate = FP / float(FP + TN)


print('False Positive Rate : {0:0.4f}'.format(false_positive_rate))

specificity = TN / (TN + FP)

print('Specificity : {0:0.4f}'.format(specificity))
```
